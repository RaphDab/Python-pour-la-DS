```python
# Cours du Nasdaq via Yahoo Finance
```

```python
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
import pandas as pd
import datetime
```

```python
NDQ = '^IXIC'
datan = yf.Ticker(NDQ) # Extraction avec yfinance
nasdaq = datan.history(period = '1d', start = '2020-1-1', end = '2023-10-26')
```

```python
nasdaq['Close'].plot()
```

```python
# Superposition avec le cours de Apple
APL = 'AAPl'
dataa = yf.Ticker(APL)
apple = dataa.history(period = '1d', start = '2020-1-1', end = '2023-10-26')
```

```python
time = np.arange(962) # Tous les jours depuis le 2020-01-01 jusqu'au 2023-10-26
```

```python
# Comparaison avec une échelle logarithmique 

plt.figure()
plt.title('Comparaison du cours du NASDAQ avec Apple')
plt.semilogy(time[800:820], apple['Close'][799:819])
plt.semilogy(time[800:820], nasdaq['Close'][799:819])
```
